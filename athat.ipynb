### 1-2. Std and var

```python
# Keep only the following columns: BasePay, OvertimePay, Benefits, TotalPay, TotalPayBenefits, Year
df2 = df[['BasePay', 'OvertimePay', 'Benefits', 'TotalPay', 'TotalPayBenefits', 'Year']]
```

```python
df2.head()
```

```python
# Convert BasePay, OvertimePay, Benefits, TotalPay, TotalPayBenefits to numeric
df2['BasePay'] = pd.to_numeric(df2['BasePay'], errors='coerce')
df2['OvertimePay'] = pd.to_numeric(df2['OvertimePay'], errors='coerce')
df2['Benefits'] = pd.to_numeric(df2['Benefits'], errors='coerce')
df2['TotalPay'] = pd.to_numeric(df2['TotalPay'], errors='coerce')
df2['TotalPayBenefits'] = pd.to_numeric(df2['TotalPayBenefits'], errors='coerce')
```

```python
from math import sqrt

def std_local(df):
    # convert df to array
    df_array = df.to_numpy()
    # remove nan values
    df_array = df_array[~np.isnan(df_array)]
    m = sum(df_array) / df_array.size
    array = [(i-m)**2 for i in df_array]
    return sqrt(sum(array) / len(array))
```

```python
print(std_local(df2['BasePay']), '=', df2['BasePay'].std())
print(std_local(df2['OvertimePay']), '=', df2['OvertimePay'].std())
print(std_local(df2['Benefits']), '=', df2['Benefits'].std())
print(std_local(df2['TotalPay']), '=', df2['TotalPay'].std())
print(std_local(df2['TotalPayBenefits']), '=', df2['TotalPayBenefits'].std())
```

```python
def var_local(df):
    # convert df to array
    df_array = df.to_numpy()
    # remove nan values
    df_array = df_array[~np.isnan(df_array)]
    m = sum(df_array) / df_array.size
    array = [(i-m)**2 for i in df_array]
    return sum(array) / len(array)
```

```python
print(var_local(df2['BasePay']), '=', df2['BasePay'].var())
print(var_local(df2['OvertimePay']), '=', df2['OvertimePay'].var())
print(var_local(df2['Benefits']), '=', df2['Benefits'].var())
print(var_local(df2['TotalPay']), '=', df2['TotalPay'].var())
print(var_local(df2['TotalPayBenefits']), '=', df2['TotalPayBenefits'].var())
```

### 3. Cv and range (using Numpy to boost efficiency)

```python
# Calculate coefficient of variations
import numpy as np

def cv_local(df):
    # convert df to array
    df_array = df.to_numpy()
    # remove nan values
    df_array = df_array[~np.isnan(df_array)]
    m = np.mean(df_array)
    if m == 0:
        return np.nan
    # calculate standard deviation with Bessel's correction (ddof=1)
    # also using NumPy functions because it's faster than local functions since core is written in C 
    std_dev = np.std(df_array, ddof=1)
    # calculate coefficient of variation and convert to percentage
    cv = (std_dev / m) * 100
    return cv
```

```python
print(cv_local(df2['BasePay']), '=', df2['BasePay'].std() / df2['BasePay'].mean() * 100)
print(cv_local(df2['OvertimePay']), '=', df2['OvertimePay'].std() / df2['OvertimePay'].mean() * 100)
print(cv_local(df2['Benefits']), '=', df2['Benefits'].std() / df2['Benefits'].mean() * 100)
print(cv_local(df2['TotalPay']), '=', df2['TotalPay'].std() / df2['TotalPay'].mean() * 100)
print(cv_local(df2['TotalPayBenefits']), '=', df2['TotalPayBenefits'].std() / df2['TotalPayBenefits'].mean() * 100)
```

```python
def range_local(df):
    # convert df to array
    df_array = df.to_numpy()
    # remove nan values
    df_array = df_array[~np.isnan(df_array)]
    return max(df_array) - min(df_array)
```

```python
print(range_local(df2['BasePay']))
print(range_local(df2['OvertimePay']))
print(range_local(df2['Benefits']))
print(range_local(df2['TotalPay']))
print(range_local(df2['TotalPayBenefits']))
```

### 4. Measures of central tendency

```python
# measures of central tendency
def ct_local(df):
    # convert df to array
    df_array = df.to_numpy()
    # remove nan values
    df_array = df_array[~np.isnan(df_array)]
    # calculate mean
    mean = sum(df_array) / df_array.size
    # calculate median
    # median = np.median(df_array)
    # if df_array.size % 2 == 0:
        #median = (df_array[df_array.size // 2] + df_array[df_array.size // 2 - 1]) / 2
    # else:
        # median = df_array[df_array.size // 2]
    # calculate mode
    mode = max(set(df_array), key = list(df_array).count)
    return mean, mode
```

```python
print(ct_local(df2['BasePay']))
print(ct_local(df2['OvertimePay']))
print(ct_local(df2['Benefits']))
print(ct_local(df2['TotalPay']))
print(ct_local(df2['TotalPayBenefits']))
```

```python
# Histogram of BasePay column
def plot_histogram(df, column, bins):
    df.plot(kind='hist', bins=bins, figsize=(10, 10))
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.show()

# Histograms where x axis is the year and y axis is the column
def plt_years(df, column, bins):
    df.hist(column=column, by='Year', bins=bins, figsize=(10, 10))
```

### 8. IQR

```python
# Calculate the IQR for the BasePay column
def iqr_local(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    return Q3 - Q1
```

```python
print(iqr_local(df2, 'BasePay'))
print(iqr_local(df2, 'OvertimePay'))
print(iqr_local(df2, 'Benefits'))
print(iqr_local(df2, 'TotalPay'))
print(iqr_local(df2, 'TotalPayBenefits'))
```
